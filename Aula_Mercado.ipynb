```python
# o nosso desafio é mostrar os produtos do Mercado
# poder escolher os produtos
# adicionar novos produtos
#quero poder excluir um produto do mercado

# criar as variaveis globais para o software
lista_produtos = ['detergente ype',
                'refrigerante Gengibirra',
                'arroz',
                'Feijão',
                'Agua Mineral']
lista_preco = [2.50 , 5, 7.25, 4, 3.45]
```

```python
# Para criarmos uma função, devemos armazenar eles dentro de um bloco.E nomear

#função 1 ver produtos
def ver_produtos() :
  print(f'Os produtos que temos são: {lista_produtos} ')
  print(f'e os seus preços são: {lista_preco}')


def adicionar_produtos() :
  novo_produto = input('Informe o nome do produto: ')
  novo_preco = float(input('Informe o preço: '))
  lista_produtos.append(novo_produto)
  # chamamos a lista e atribuimos a função append e adiciona o que vai ser adicionado
  lista_preco.append(novo_preco)
```

```python
print('-----Menu do Mercado-----')
print('escolha a opção: ')
print('1- ver produtos')
print('2- cadastrar novo produto')
opcao = int(input('Informe a opção desejada'))

# comando IF para condicionar
if opcao ==1:
  #o que ele deve executar
  ver_produtos()
# elif
elif opcao ==2:
  adicionar_produtos()
  ver_produtos()
else:
   print('opcao não encontrada')
```
